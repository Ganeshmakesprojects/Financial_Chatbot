# Finance Data
Collected the fiancial data for 3 companies (Tesla, Apple and Microsoft) for last 3 years 2023,2022 and 2021.

```python
df = pd.read_excel("Companies_3year_data.xlsx", index_col="Company")
df = df.sort_values(by=['Company', 'Year'], ascending=True)
df
```

```python
"Total Revenue of Apple: {}".format(df[(df.Year == 2023)].loc["Apple", "Net Income"])
```

# Initial Observations

Apple is bigger company of the 3 w.r.t Revenue, Income and cash with Tesla being the smallest among the 3.

However, Microsoft has better assets and cash w.r.t to size of the companies.

```python
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income'].pct_change() * 100
df.fillna(0, inplace=True)
df
```

# Revenue & Income Growth
Tesla has the most growth among the 3 companies.

All the 3 companies had their best growth in 2022 and slow down in 2023 which also reflects in cashflow.

```python
df['Profit Margin'] = df['Net Income']/df['Total Revenue']*100
df['Profit Margin Growth (%)'] = df.groupby(['Company'])['Profit Margin'].pct_change() * 100
df
```

# Profit Margin

Microsoft has the best profit margin, Apple with healthy margin and Tesla with lowest margins in mid teens.

However, Tesla has significant growth in its margins with Apple maintaining it's margins over the 2 years. But, micorsoft has slight reduction in profit margin.
